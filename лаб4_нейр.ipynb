```python
import pandas as pd

# Загрузка и предобработка данных
df = pd.read_csv('Mall_Customers.csv')
del df['CustomerID']

# Замена значений 'Male' и 'Female' на 1 и 2 соответственно
df['Gender'] = df['Gender'].replace({'Male': 1, 'Female': 2})

# Преобразование DataFrame в массив списков
data_array = df.values.tolist()
```

```python
print(data_array)
```

Описываем модель и проводим моделирование. Зададим в качестве начального значения количество кластеров равное трем.

Этот код создает модель кластеризации методом k-средних (KMeans) с указанием количества кластеров равным 3 (n_clusters=3). Затем модель обучается на данных, представленных в data_array, чтобы найти оптимальные центроиды для каждого кластера, что позволяет разделить данные на 3 кластера.

```python
model = KMeans(n_clusters=3)
model.fit(data_array)
```

Получим предсказание на единичном примере и выведем результат на экран.

```python
predicted_label = model.predict([[2, 35, 8, 16]])
print(predicted_label)
```

Получим также предсказание на всем наборе данных и выведем результаты на экран.

```python
all_predictions = model.predict(data_array)
print(all_predictions)
```

Представление в формате простых чисел не совсем наглядно, поэтому попробуем графически представить результаты кластеризации.

```python
for i in range(0, len(data_array)):
    if all_predictions[i] == 0:
        c1 = plt.scatter(data_array[i][1], data_array[i][0], c='r', marker='+')
    elif all_predictions[i] == 1:
        c2 = plt.scatter(data_array[i][1], data_array[i][0], c='g', marker='o')
    elif all_predictions[i] == 2:
        c3 = plt.scatter(data_array[i][1], data_array[i][0], c='b', marker='*')

plt.legend([c1, c2, c3], ['Кластер 1', 'Кластер 2', 'Кластер 3'])
plt.title('k-means. Разделение на три кластера')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score,v_measure_score
scores= []
res_silhouette = []
vmeasure_score =[]
db_score = []
for i in range(2,12):
    model = KMeans(n_clusters=i, random_state=0).fit(data_array)
    preds = model.predict(data_array)
    #print("Score for number of cluster(s) {}: {}".format(i,model.score(data_array)))
    scores.append(-model.score(data_array))
    silhouette = silhouette_score(data_array,preds)
    res_silhouette.append(silhouette)
    #print("Silhouette score for number of cluster(s) {}: {}".format(i,silhouette))

plt.figure(figsize=(7,4))
plt.title("The elbow method for determining number of clusters\n",fontsize=16)
plt.scatter(x=[i for i in range(2,12)],y=scores,s=150,edgecolor='k')
plt.grid(True)
plt.xlabel("Number of clusters",fontsize=14)
plt.ylabel("K-means score",fontsize=15)
plt.xticks([i for i in range(2,12)],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.title("The silhouette coefficient method \nfor determining number of clusters\n",fontsize=16)
plt.scatter(x=[i for i in range(2,12)],y=res_silhouette,s=150,edgecolor='k')
plt.grid(True)
plt.xlabel("Number of clusters",fontsize=14)
plt.ylabel("Silhouette score",fontsize=15)
plt.xticks([i for i in range(2,12)],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```
